### GFlowNets clustering small synthetic graphs
This file shows how to use the GFlowNet for clustering small graphs generating using the Infinite Relational Model (IRM). With graphs of < 7 nodes the computations are manageable within a reasonable timeframe, as the following code calculates the full posterior distribution. See the function 'compare_results_small_graphs' for running multiple examples and generating their plots all at once.

```python
import torch
import numpy as np
from tqdm import tqdm
from Core.Core import IRM_graph, GraphNet, GraphNetNodeOrder, allPosteriors, empiricalSampleDistribution, Gibbs_sample_torch, plot_posterior
```

To generate the graphs the priors for the IRM model and the number of nodes in the synthetic graph must be set. Here we choose a graph with 5 nodes and flat priors of 1

```python
N = 5
a = 1.
b = 1.
A_alpha = 1.
```

Next we must choose whether the network uses a fixed but randomly initialised node order for ever time it clusters the graph or gets to choose the order itself. Choosing the node order costs O(x^2) whereas the randomly initialised costs O(x) for each pass through the graph, where x is the number of nodes in the graph. With this chosen, the following creates a graph using the IRM process and initialises a GFlowNet.

```python
useFixedNodeOrder = False
adjacency_matrix, clusters = IRM_graph(A_alpha=A_alpha, a=a, b=b, N=N)
net = GraphNet(nNodes=adjacency_matrix.size()[0], a=a, b=b, A_alpha=A_alpha) if not useFixedNodeOrder else GraphNetNodeOrder(
                nNodes=adjacency_matrix.size()[0], a=a, b=b, A_alpha=A_alpha)
```

Now the full posterior distribution is calculated

```python
cluster_post = allPosteriors(adjacency_matrix, a, b, A_alpha, log=True, joint=False)
```

Next the number of samples for the empirical distribution must be specified, along with the number of epochs of training and the intervals at which to draw new samples for training.

```python
max_epochs = 10
epoch_interval = 5
nSamplesDistribution = 1000

for epochs in tqdm(range(0, max_epochs, epoch_interval), desc='Epoch iteration'):
    X1 = net.sample_forward(adjacency_matrix, nSamples=nSamplesDistribution, timer=True)
    sample_posteriors_numpy = empiricalSampleDistribution(X1, N, log=True, numpy=True, minVal=np.min(cluster_post))
    sort_idx = np.argsort(cluster_post)
    net.train(X1, epochs=epoch_interval, verbose=False)
```

Finally, the results are visualised in a plot comparing the true posterior IRM values with the empirical distributions from the Gibbs sampler and the GFlowNet. The plot is displayed and saved with a filename containing the number of nodes, epochs of training and number of samples drawn for the distribution. It must be noted that clusterings never sampled empirically are set to being seen once.

```python
gibbsSamples = Gibbs_sample_torch(adjacency_matrix, nSamplesDistribution * 2, return_clustering_matrix=True)
tempSamples = torch.zeros((nSamplesDistribution, N ** 2 * 2))
for i, sample in enumerate(gibbsSamples):
    tempSamples[i] = torch.concat((adjacency_matrix.flatten(), torch.tensor(sample.flatten())))
gibbsDistribution = empiricalSampleDistribution(tempSamples, N, numpy=True, log=True, minVal=np.min(cluster_post))

plot_posterior(cluster_post,
               sort_idx=sort_idx,
               net_posteriors_numpy=None,
               gibbs_sample_posteriors=gibbsDistribution,
               sample_posteriors_numpy=sample_posteriors_numpy,
               saveFilename=f'Plots/PosteriorPlotLog_{N}_{max_epochs}_{nSamplesDistribution}.png',
               title=f'Posterior values from GflowNet, Gibbs and IRM\nOn graph of {N} nodes')
```

This plot shows the log posterior probability of each cluster. They are ordered in ascending order from left to right according to their true posterior values. To get the non-log values, the values are exponentiated:

```python
plot_posterior(np.exp(cluster_post),
               sort_idx=sort_idx,
               net_posteriors_numpy=None,
               gibbs_sample_posteriors=np.exp(gibbsDistribution),
               sample_posteriors_numpy=np.exp(sample_posteriors_numpy),
               saveFilename=f'Plots/PosteriorPlot_{N}_{max_epochs}_{nSamplesDistribution}.png',
               title=f'Posterior values from GflowNet, Gibbs and IRM\nOn graph of {N} nodes')
```
